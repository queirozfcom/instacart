## building a denormalized prior set

the denormalized dataset should have one row for each product

i know this will be huge and wasteful, but this is just meant to 
help me think about what structure I need the data to be in

this dataset will have data from BOTH the "train" as from the "prior" datasets

```python
prior_df.describe()
```

first off - see what eval_set the users are in

```python
prior_df_order_ids  = prior_df["order_id"]
```

```python
prior_df_user_ids = prior_df_order_ids.apply(lambda order_id: orders_df.iloc[])
```

now for the actual denormalization

```python
# add order information
prior_df = pd.concat([prior_df,prior_df.apply(lambda row: orders_df.iloc[row['order_id']],axis=1)],axis=1)
prior_df.sample(5)
```

```python
# add product information
prior_df= pd.concat([prior_df, prior_df.apply(lambda row: products_df.iloc[row['product_id']], axis=1)],axis=1)
prior_df.sample(5)
```

```python
# add aisle information
prior_df= pd.concat([prior_df, prior_df.apply(lambda row: aisles_df.iloc[row['aisle_id']], axis=1)],axis=1)
prior_df.sample(5)
```

```python
# add department information
prior_df= pd.concat([prior_df, prior_df.apply(lambda row: departments_df.iloc[row['department_id']], axis=1)],axis=1)
prior_df.sample(5)
```

```python
prior_df.drop(['product_id','aisle_id','department_id'],inplace=True,axis=1)
```

```python
prior_df.sample(5)
```

## what users are in prior_df but not in train_df?

(these are the ones that we have to predict the items for)

```python
unique_order_ids_in_prior_df = prior_df["order_id"].unique()
```

```python
def maybe_get_user_id(order_id):
    try:
        user_id = orders_df.iloc[order_id]["user_id"]
    except IndexError:
        return None
    
    return user_id
```

```python
user_ids_in_prior_df = [maybe_get_user_id(order_id) for order_id in unique_order_ids_in_prior_df ]
user_ids_in_prior_df = [ user_id for user_id in user_ids_in_prior_df if user_id is not None]
```

```python
user_ids_in_prior_df = set(user_ids_in_prior_df)
```

```python
len(user_ids_in_prior_df)
```

```python
unique_order_ids_in_train_df = train_df["order_id"].unique()
```

```python
user_ids_in_train_df = [maybe_get_user_id(order_id) for order_id in unique_order_ids_in_train_df ]
user_ids_in_train_df = [ user_id for user_id in user_ids_in_train_df if user_id is not None]
user_ids_in_train_df = set(user_ids_in_train_df)
```

```python
len(user_ids_in_train_df)
```

```python
user_ids_in_prior_df
```

```python
user_ids_in_train_df
```

```python
user_ids_to_test = user_ids_in_prior_df - user_ids_in_train_df
```

```python
len(user_ids_to_test)
```
